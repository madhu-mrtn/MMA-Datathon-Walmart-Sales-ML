```python
import pandas as pd
```

```python
data=pd.read_csv('refrigerated_df.csv')
```

```python
produce_data = data[data['department']=='produce']
produce_products = produce_data['product_name'].unique()
```

```python
len(produce_products)
```

```python
from fuzzywuzzy import fuzz, process

# Load the data into a DataFrame
data = produce_data

# Step 2: Calculate sales for each product
product_sales = data.groupby('product_name')['order_id'].nunique().reset_index()
product_sales.columns = ['product_name', 'sales']

# Step 3: Group similar products using fuzzy string matching
product_groups = {}
threshold = 50  # You can adjust the threshold as needed

unique_products = product_sales['product_name'].unique()

for product in unique_products:
    similar_products = process.extractBests(product, unique_products, scorer=fuzz.token_set_ratio, score_cutoff=threshold)
    if similar_products:
        group = [sim_product for sim_product, score in similar_products]
        max_sales_product = max(group, key=lambda x: product_sales[product_sales['product_name'] == x]['sales'].values[0])
        product_groups[max_sales_product] = group

# Step 4 and 5: Select products with the highest sales in each group and add them to the basket
basket = list(product_groups.keys())

```
